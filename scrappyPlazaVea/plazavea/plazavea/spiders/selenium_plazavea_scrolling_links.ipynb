```python
from selenium import webdriver
from time import sleep
import pandas as pd
import numpy as np

driver = webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.plazavea.com.pe/laptops?ft=laptops") #Link raiz
sleep(6)
#print(driver.find_elements_by_xpath("/html/body/div[4]/main/div/div[3]/div/section[2]/div[1]/div[3]/span[1]")[0].text)
# elements_total = int(driver.find_elements_by_xpath("/html/body/div[4]/main/div/div[3]/div/section[2]/div[1]/div[3]/span[1]")[0].text) #Numero de productos encontrados
elements_total = 338
elements_actual = len(driver.find_elements_by_class_name("Showcase__link")) #Es una lista q contiene todos los links de productos

container = driver.find_elements_by_class_name("Showcase__link")

print('Mostrando' ,str(elements_actual), 'productos de un total de', str(elements_total))

while elements_actual < elements_total:
    driver.execute_script("arguments[0].scrollIntoView();",container[elements_actual-1])
    sleep(5)
    elements_actual = len(driver.find_elements_by_class_name("Showcase__link"))
    print('Mostrando' ,str(elements_actual), 'productos de un total de', str(elements_total))
    
    container = driver.find_elements_by_class_name("Showcase__link")
    
    if elements_actual == elements_total:
        break

driver.execute_script("arguments[0].scrollIntoView();", container[elements_actual-1])
print('--> Links extraídos')

#Guardando en un CSV
links = []
for i in container:
    mm = i.get_attribute("href")
    links.append(mm)
    
df = pd.DataFrame({'links':links}) 
df.to_csv('links.csv', index = False) 
print('------------------------')
print('Archivo *csv descargado')
```

## Código para extracción de precios de cada web de Plaza vea

```python
csv_links = pd.read_csv("links.csv")
list_links =list(csv_links['links'])
```

```python
#Guardando en un CSV
id = []
precio_normal_list = []
precio_online_list = []
precio_oh_list = []

for i in range(len(list_links)-1):
    driver.get(list_links[i])
    sleep(4)
    # print("linea i=","i ",list_links[i])

    try:
        precio_normal = driver.find_elements_by_xpath("/html/body/div[4]/div[4]/div[1]/div[4]/div[2]/div[3]/div[2]/div/span[2]")[0].text
    except:
        precio_normal = None

    try:
        precio_online = driver.find_elements_by_xpath("/html/body/div[4]/div[4]/div[1]/div[4]/div[2]/div[3]/div[3]/div[1]/span[2]")[0].text
    except:
        precio_online = None    

    try:    
        precio_oh = driver.find_elements_by_xpath("/html/body/div[4]/div[4]/div[1]/div[4]/div[2]/div[3]/div[4]/div[1]/span[2]")[0].text
    except:
        precio_oh = None

    id.append(i+1)
    precio_normal_list.append(precio_normal.split())
    precio_online_list.append(precio_online.split())
    precio_oh_list.append(precio_oh.split())
    
```

```python
# convertir a un dataframe todas las listas
df = pd.DataFrame(list(zip(id,list_links,precio_normal_list,precio_online_list,precio_oh_list)), 
                    columns = ['id','link','precio_normal','precio_online','precio_oh'] )
df.to_csv('id_precio.csv', index = False) 
print('------------------------')
print('Archivo id_precio.csv descargado')
```
